```python
import random

# Function to simulate checking an order in the database
def check_order_status(order_number):
    # Simulate an order if it doesn't exist
    shipped_date = "2026-02-" + str(random.randint(20, 28))
    estimated_delivery = "2026-03-" + str(random.randint(1, 5))
    tracking_number = "FSX" + str(random.randint(100000000, 999999999))
    carrier = "FastShip Express"
    status = random.choice(["Processing", "Shipped", "In Transit", "Delivered"])

    print(f"\nOrder #{order_number} Status:")
    print(f"  Status: {status}")
    print(f"  Shipped Date: {shipped_date}")
    print(f"  Estimated Delivery: {estimated_delivery}")
    print(f"  Carrier: {carrier}")
    print(f"  Tracking Number: {tracking_number}")
    print(f"  Tracking Link: https://www.fastship.com/track/{tracking_number}")

    # Return the final status for the single update
    return status

def subscribe_updates(order_number, status):
    print(f"\nYou are now subscribed to updates for order #{order_number}.")
    print("Final delivery update:")
    if status in ["Processing", "Shipped", "In Transit"]:
        print(f"Your package is currently '{status}' and expected to arrive soon.")
    else:
        print(f"Your package has been '{status}'.")

def main():
    print("Welcome to BrightCart Customer Support AI!")
    order_number = input("Please enter your order number: ").strip()

    # Check order status and get current status
    status = check_order_status(order_number)

    subscribe = input("\nWould you like to receive a delivery update? (yes/no): ").strip().lower()
    if subscribe == "yes":
        subscribe_updates(order_number, status)
    else:
        print("No updates subscribed. Thank you!")

if __name__ == "__main__":
    main()
```
