```python
# Text Analysis Program

import string
from collections import Counter

def analyze_text(text):
    # Safety for empty input
    if not text.strip():
        print("Total Words: 0")
        print("Total Sentences: 0")
        print("Most Used Word: None")
        print("Frequency: 0")
        return

    # Count sentences by ., !, ?
    sentence_endings = ".!?"
    sentence_count = sum(text.count(c) for c in sentence_endings)

    # Normalize text: lowercase and replace punctuation with spaces
    translator = str.maketrans(string.punctuation, " " * len(string.punctuation))
    clean_text = text.translate(translator).lower()

    # Split into words
    words = clean_text.split()
    total_words = len(words)

    # Handle no words case
    if total_words == 0:
        most_used_word = None
        frequency = 0
    else:
        word_counts = Counter(words)
        most_used_word, frequency = word_counts.most_common(1)[0]

    # Print results
    print(f"Total Words: {total_words}")
    print(f"Total Sentences: {sentence_count}")
    print(f"Most Used Word: {most_used_word}")
    print(f"Frequency: {frequency}")


# Allow user to paste text input
user_text = input("Paste your text here: ")
analyze_text(user_text)
```
