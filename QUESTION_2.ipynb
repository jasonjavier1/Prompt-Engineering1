```python
# Python program to analyze text (improved version)
def analyze_text(text):
    import re
    from collections import Counter

    # Handle empty input
    if not text or text.strip() == "":
        return {
            "total_words": 0,
            "total_sentences": 0,
            "most_used_word": None,
            "frequency": 0
        }

    # Normalize text to lowercase
    text_lower = text.lower()

    # Replace all non-alphanumeric characters (except sentence delimiters) with space
    text_for_words = re.sub(r"[^a-z0-9\s]", " ", text_lower)

    # Split words and filter out empty strings
    words = [word for word in text_for_words.split() if word]
    total_words = len(words)

    # Count word frequency
    word_counts = Counter(words)
    if word_counts:
        most_used_word, frequency = word_counts.most_common(1)[0]
    else:
        most_used_word, frequency = None, 0

    # Count sentences using regex to split on one or more of ., !, ?
    sentences = re.split(r'[.!?]+', text)
    sentences = [s.strip() for s in sentences if s.strip()]  # Remove empty strings
    total_sentences = len(sentences)

    return {
        "total_words": total_words,
        "total_sentences": total_sentences,
        "most_used_word": most_used_word,
        "frequency": frequency
    }

# --- User Input ---
user_text = input("Paste your text here:\n")

# Analyze text
result = analyze_text(user_text)

# Print results clearly
print(f"Total Words: {result['total_words']}, "
      f"Total Sentences: {result['total_sentences']}, "
      f"Most Used Word: {result['most_used_word']}, "
      f"Frequency: {result['frequency']}")
```
